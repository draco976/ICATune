```python
import matplotlib.pyplot as plt 
import numpy as np
import pickle
```

```python
acc_8b = pickle.load(open("dump/standard_acc.pkl", "rb"))
acc_3b = pickle.load(open("dump/standard_acc_3B.pkl", "rb"))
acc_1b = pickle.load(open("dump/standard_acc_1B.pkl", "rb"))

acc_8b_d = pickle.load(open("dump/disambiguated_acc.pkl", "rb"))
acc_3b_d = pickle.load(open("dump/disambiguated_acc_3B.pkl", "rb"))
acc_1b_d = pickle.load(open("dump/disambiguated_acc_1B.pkl", "rb"))

acc_8b_d_red = pickle.load(open("dump/disambiguated_acc_reduced.pkl", "rb"))
acc_3b_d_red = pickle.load(open("dump/disambiguated_acc_3B_reduced.pkl", "rb"))
acc_1b_d_red = pickle.load(open("dump/disambiguated_acc_1B_reduced.pkl", "rb"))
```

```python
c_values = np.arange(1, 9)

# Bar width
bar_width = 0.2

# Bar positions
x = np.arange(len(c_values))
model1_positions = x - bar_width
model2_positions = x
model3_positions = x + bar_width

# Plot
plt.figure(figsize=(9, 3))
plt.bar(model1_positions, acc_1b[:, 7], width=bar_width, label="Llama-3.2-1B")
plt.bar(model2_positions, acc_3b[:, 7], width=bar_width, label="Llama-3.2-3B")
plt.bar(model3_positions, acc_8b[:, 7], width=bar_width, label="Llama-3.1-8B")

# Labels and title
plt.xlabel("Length of y-phrase ($c$)", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
# plt.title("Predicion Accuracy Comparison for C=1 to C=8", fontsize=14)
plt.xticks(x, c_values)
plt.legend()
plt.tight_layout()

plt.savefig("figures/compare_models.png", dpi=300, bbox_inches='tight')
```

```python
c_values = np.arange(1, 9)

# Bar width
bar_width = 0.2

# Bar positions
x = np.arange(len(c_values))
model1_positions = x - bar_width
model2_positions = x
model3_positions = x + bar_width

# Plot
plt.figure(figsize=(9, 3))
plt.bar(model1_positions, acc_1b_d[:, 7], width=bar_width, label="Llama-3.2-1B")
plt.bar(model2_positions, acc_3b_d[:, 7], width=bar_width, label="Llama-3.2-3B")
plt.bar(model3_positions, acc_8b_d[:, 7], width=bar_width, label="Llama-3.1-8B")

# Labels and title
plt.xlabel("Length of y-phrase ($c$)", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
# plt.title("Predicion Accuracy Comparison for C=1 to C=8", fontsize=14)
plt.xticks(x, c_values)
plt.legend(loc='lower left')
plt.tight_layout()

plt.savefig("figures/compare_models_d.png", dpi=300, bbox_inches='tight')
```

```python
# Data setup
models = ["8B", "3B", "1B"]
standard = [acc_8b[0,7], acc_3b[0,7], acc_1b[0,7]]
disambiguated = [acc_8b_d[0,7], acc_3b_d[0,7], acc_1b_d[0,7]]
reduced = [acc_8b_d_red[0,7], acc_3b_d_red[0,7], acc_1b_d_red[0,7]]

x = np.arange(len(models))  # Bar positions
width = 0.2  # Bar width

# Create plot
fig, ax = plt.subplots(figsize=(7, 4))
rects1 = ax.bar(x - width, standard, width, label='Standard')
rects2 = ax.bar(x, disambiguated, width, label='Pre-Aligned')
rects3 = ax.bar(x + width, reduced, width, label='Reduced Pre-Aligned')

# Labels and formatting
ax.set_xlabel("Model Size", fontsize=12)
ax.set_ylabel("Accuracy", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(loc='lower right')

plt.savefig("figures/compare_models_disambiguated_reduced.png", dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
